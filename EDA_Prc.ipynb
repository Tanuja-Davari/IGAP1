```python
#Importing matplotlib library
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Line Plot
x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("data 1")
plt.ylabel("data 2")
plt.show()
```

```python
#Bar Plot
import pandas as pd
data = pd.DataFrame({
    "class":['A','B','C','D','E'],
     "Students":[50,40,55,36,70]
})
sns.barplot(x="class",y="Students",data=data)#
plt.title("Bar-Plot")
plt.xlabel("Classes")
plt.ylabel("Students")
plt.show()
```

```python
#Count Plot
import pandas as pd
data = pd.DataFrame({
    "class":['A','B','C','D','E'],
     "Students":[50,40,55,36,70]
})
sns.countplot(x='class',data=data#Count no.of rows
plt.title("Count Plot")
plt.show()
```

```python
#Histogram
import numpy as np
data=np.random.randn(100)
plt.hist(data,bins=10)#Data Distribution
plt.title("Histogram")
plt.show()
```

```python
#KDE Plot
import numpy as np
data=np.random.randn(100)
sns.kdeplot(data,shade=True)#Show distribution of data
plt.title("KDE Plot")
plt.show()
```

```python
#Histogram with KDE
import numpy as np 
sns.histplot(data,bins=10,kde=True)
plt.title("Histogram with KDE")
plt.show()
```

```python
#Box Plot
import numpy as np
sns.boxplot(y=data)#Show outliers
plt.title("Box Plot")
plt.show()
```

```python
#Violin Plot
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
data=np.random.randn(100)
sns.violinplot(y=data)#Show combination of box lot and KDE
plt.title("violin plot")
plt.show()
```

```python
#Scatter Plot
x=[1,2,3,4,5]
y=[5,4,7,10,8]
plt.scatter(x,y)#show relationship between two variables
plt.title("Scatter Plot")
plt.show()
```

```python
#Pair Plot
#Loading dataset
df=sns.load_dataset('iris')#Show relationship between all dataset frames
print(df)
```

```python
#Plotting pair plot
import matplotlib.pyplot as plt
sns.pairplot(df,hue='species')
#plt.shoW()
```

```python
#HeatMap
df_numeric=df.drop(columns='species')
df_numeric=df.drop('species',axis=1)
correlation_matrix=df_numeric.corr()#Show visulizations matrix and correlation
correlation_matrix
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("Co-relation Heatmap")
```

```python
#Pie-Chart
labels=['A','B','C','D']
sizes=[30,50,20,10]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')#Relationship between categorical and numerical variable
plt.title("Pie-Chart")
plt.show()
```

```python
#Swarm Plot
sns.swarmplot(x='species',y='sepal_length',data=df)#Show range of data that spreaded 
plt.title("Swarm Plot")
plt.show()
```

```python
#Strip Plot
sns.stripplot(x='species',y='sepal_length',data=df,jitter=True)
plt.title("Strip Plot")
plt.show()
```

```python
#Joint Plot
import seaborn as sns
sns.jointplot(x='sepal_length',y='petal_length',data=df,kind='scatter')#Show combination of Scatter and Histogram
plt.show()
```
