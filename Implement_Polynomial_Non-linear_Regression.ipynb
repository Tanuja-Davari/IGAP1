```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing and reading dataset
ds = pd.read_csv("Position_Salaries.csv")
ds
```

```python
#Divide or split dataset into Independent and Dependent data
x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
#Plot graphs
plt.scatter(x,y)#Ploynomial(curved) get from it i.e,non-linear
plt.show()
```

```python
#Model fitting (Polynomial regression model fitting)
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures#To create line from curved
poly_reg = PolynomialFeatures(degree=5)#Degree given by default it is 5
x_poly = poly_reg.fit_transform(x)
#x_poly
reg = LinearRegression()#Convet non-linear into linear regression
reg.fit(x_poly,y)#To fit model
```

```python
#Visulization to show how fit that line
plt.scatter(x,y)
plt.plot(x,reg.predict(poly_reg.transform(x)),color='green')
plt.show()
```

```python
#Predicting for myinput(For manually predict)
myinput = pd.DataFrame(data=[[11]],columns=['Level'])
result = reg.predict(poly_reg.transform(myinput))
result
```
