```python
#Making new dataframe using categorical columns only to recommend movies
new_df = merge_df[['id','genres','keywords','overview','title_x','cast','crew','production_companies']]
new_df.head(2)
```

```python
#Changing column heading
new_df = new_df.rename(columns={'title_x':'title'})#Rename col title_x with title
```

```python
#Using function convert string to comma separated values

import ast#
def get_names(obj):
    x = ast.literal_eval(obj)#get string values in dict
    keywords=[]
    for r in x:
        name = r["name"].replace(" ","_").lower().strip()#It take only string values after name col only
        if name not in keywords:
            keywords.append(name)
    return",".join(keywords)
```

```python
new_df["keywords"]=new_df["keywords"].apply(get_names)#Convert dict into comma seprated values 
```

```python
new_df
```

```python
new_df["genres"] = new_df["genres"].apply(get_names)
```

```python
new_df["cast"]=new_df["cast"].apply(get_names)
```

```python
new_df["crew"] = new_df["crew"].apply(get_names)
```

```python
new_df["production_companies"]=new_df["production_companies"].apply(get_names)
```

```python
new_df['overview'] = new_df['overview'].apply(lambda x:str(x).replace(",","").replace(" ",",").lower())
```

```python
new_df["tags"] = new_df["genres"]+","+new_df["keywords"]+"," +new_df["overview"]+"," +new_df["cast"]+"," +new_df["production_companies"]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#Download stopwords if not already available
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def remove_stop_words(text):
    #get the english stopwords list
    stop_words = set(stopwords.words('english'))
    #Tokenize the sentense
    words = text.split(',')
    #Remove stopwords
    filtered_words = [word for word in words if word.lower() not in stop_words]
    #Convert back to string 
    filtered_text = ",".join(filtered_words)
    return filtered_text
```

```python
final_df["tags"] = final_df["tags"].apply(remove_stop_words)
```
