```python
#Importing Libraries
import numpy as np 
import pandas as pd
```

```python
#Importing dataset
ds =pd.read_csv("cleaned_data.csv")
ds
```

```python
for i in ds.index:
    print(ds.loc[i,"title"],"-",ds.loc[i,"tags"].count(","))
```

```python
#bag of words (transform text to numeric format)
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features =5000,stop_words ='english')
```

```python
vector = cv.fit_transform(ds["tags"]).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
#Draw 2D array (To see similarity between movies)

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
#Getting movie index
def get_movie_index(name):
    movie_ds_index = -1
    for i in ds.index:
        if ds.loc[i,"title"].lower()==name.lower():
            movie_ds_index = i
            break
    return movie_ds_index
```
