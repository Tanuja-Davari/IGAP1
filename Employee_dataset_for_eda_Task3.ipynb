```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("employee_dataset_for_eda.csv")
ds
```

```python
ds.info()
```

```python
#Employee Count per Department
sns.countplot(data=ds,x='Department',palette='Set2')
plt.title('Employee count per depatment')
```

```python
Interpretetion : Finance Dept having more employees followed by Marketing and Engineering Dept.
```

```python
sns.countplot(data=ds,x='Gender',palette='coolwarm')
plt.title("Gender Distribution")
```

```python
Interpretetion : Count of Female more than as compared to Male.
```

```python
#Employee Location Distribution
sns.countplot(data=ds,x='Location',palette='Spectral')
plt.title("Number of Employee by  Loctaion")
plt.xticks(rotation=45)
```

```python
Interpretetion : 
```

```python
#Age Distribution
sns.histplot(ds['Age'],bins=10,kde=True)
plt.title("Age Distribution")
```

Interpretetion : Peak between 45-50 have more experienced peoples than peak between 25-30.

```python
#Years of Experience Distribution
sns.histplot(ds["Years_of_Experience"],bins=15,kde=True)
plt.title("Years of Experience")
```

```python
Interpretetion : 5.0 Years of experience is greater then another.
    
```

```python
#Monthly Salary Distribution
sns.boxplot(data=ds,y='Monthly_Salary',color='lightblue')
plt.title('Salary Spread')
```

```python
Interpretetion: 
```

```python
#Average salary of Department
sns.barplot(data=ds,x='Department',y='Monthly_Salary',estimator=np.mean,palette='viridis')
plt.title("Average Salary of Department")
```

Interpretetion : Average salary of Finance department is more than all.

```python
#Performance Score by Department
sns.boxplot(data=ds,x='Department',y='Performance_Score',palette='pastel')
plt.title("Performance Score by Department")
```

```python
Interpretetion : 
```

```python
#Salary Distribution by Gender
sns.violinplot(data=ds,x='Gender',y='Monthly_Salary',palette='muted')
plt.title("Salary Distribution by Gender")
```

```python
Interpretetion : 
```

```python
#Correlation Matrix
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")#Show negative correlation
```

```python
#Salary vs Experience Scatter Plot
sns.scatterplot(data=ds,x='Years_of_Experience',y='Monthly_Salary',hue='Department')
plt.title("Salary vs.Experience by Department")
```

```python
 Interpretetion : As Year_of_Experience is more then salary also increases.
Finance,Marketing,Engineering have more experience than HR and sales.
```
