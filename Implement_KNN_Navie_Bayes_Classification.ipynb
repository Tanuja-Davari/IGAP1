```python
#11/7

#Import Libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import and read dataset
ds = pd.read_csv("Social_Network_Ads.csv")
ds
```

```python
#Check Duplicates count
ds.duplicated().sum()
```

```python
#Remove Duplicates
ds.drop_duplicates(inplace=True)
```

```python
#Check empty cell count
ds.isnull().sum()
```

```python
#Splitting data into Independent and Dependent variable

x = ds[["Age","EstimatedSalary"]]
y = ds[["Purchased"]]
```

```python
x
```

```python
y
```

```python
# Feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
# Dividing dataset into Training and Testing data

from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
# Classification model fitting

from sklearn.neighbors import KNeighborsClassifier
cls = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)
cls.fit(x_train, y_train)
```

```python
# predict for my input

cls.predict(sc.transform([[40,97000]]))
```

```python
# Concatenate the prediction and original output

y_pred = cls.predict(X_test)
y_pred = pd.DataFrame(data = y_pred.reshape(92,1), columns = ["Prediction"])
```

```python
y_test
```

```python
result = pd.DataFrame(np.concatenate((y_test, y_pred), axis = 1), columns = ["Purchased","Prediction"])
result
```

```python
# Cheaking accuracy using confusion matrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(result["Purchased"], result["Prediction"])
cm
```

```python
score = accuracy_score(result["Purchased"], result["Prediction"])       # we get 93% accuracy in prediction
score
```

```python
# User Input

age = float(input("Enter Age: "))
salary = float(input("Enter Estimated Salary: "))

myinput = pd.DataFrame(sc.transform([[age, salary]]), columns=['Age', 'EstimatedSalary'])

result = cls.predict(myinput)

if result[0] == 1:
    print("Prediction: User will purchase plan")
else:
    print("Prediction: User will not purchase plan")
```
