```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Import dataset
import pandas as pd
ds=pd.read_csv("product_dataset_Task2.csv")
ds.head(5)#only top of 5 dataset show
```

```python
print(ds.to_string())
```

```python
ds.info()
```

```python
ds.describe()#describe all info such as count,unique
```

```python
#Searching for duplicates
ds.duplicated().sum()#return count of duplicates
```

```python
#Remove duplicates
ds.drop_duplicates(inplace = True)#return count of duplicate 0
```

```python
#Name of columns(headers)
ds.columns
```

```python
#Handling Price
ds["Price"]=pd.to_numeric(ds["Price"],errors='coerce')#age convert into float
ds.info()
```

```python
ds["Price"].unique()
```

```python
#TO fill missing values
x=ds["Price"].median()
ds["Price"].fillna(x,inplace = True)#Fill the all missing values in column price
ds.info()
```

```python
print(ds.to_string())
```

```python
ds["Quantity"].unique()
```

```python
#Stanerdized City
ds["Quantity"] = ds["Quantity"].replace({
    "two":"4.0","1.0":"5.0","":"None"
})
ds.info()
```

```python
print(ds.to_string())
```

```python
ds.columns
```

```python
#Handling Quantity Column
ds["Quantity"].fillna("-",inplace = True)
ds.info()
```

```python
print(ds.to_string())
```

```python
#Handling Purchasedate column
ds["PurchaseDate"]=pd.to_datetime(ds["PurchaseDate"],errors='coerce')#Dtype of Join_date convert into datetime
ds.info()
```

```python
#Handling Quantity column
ds["Quantity"]=pd.to_numeric(ds["Quantity"],errors='coerce')#Dtype of Quantity convert object into float
ds.info()
```

```python
#unique values in Department
ds["CustomerID"].unique()
```

```python
#Mode of col CustomerID
y=ds["CustomerID"].mode()[0]
y
```

```python
#Handling misssing values of CustomerID column
ds["CustomerID"].fillna(y,inplace=True)
print(ds.to_string())
```

```python
#Save data into new file
ds.to_csv("product_dataset_Task2.csv")
```
