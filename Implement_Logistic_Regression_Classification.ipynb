```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
#Importing and reading dataset
ds = pd.read_csv("Social_Network_Ads.csv")
ds
```

```python
ds.duplicated().sum()#TO check count of duplicates
```

```python
ds.drop_duplicates(inplace=True)#Remove Duplicates
```

```python
ds.isnull().sum()#To check count of empty cells
```

```python
#Splitting as Independent and Dependent variables
x=ds[['Age','EstimatedSalary']]
y=ds[['Purchased']]
```

```python
x
```

```python
y
```

```python
#Feature Scaling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = pd.DataFrame(sc.fit_transform(x),columns = ['Age','EstimatedSalary'])#To convert larger datavalue into smaller format
```

```python
#Splitting the dataset into training and test set
from sklearn.model_selection import train_test_split#Split vertically using train_test_split()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1,random_state = 0)
```

```python
len(x_test)
```

```python
#Applying the Logistic Regression model on the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
#Predict for myinput
myinput = pd.DataFrame(sc.transform([[30,150000]]),columns = ['Age','EstimatedSalary'])
classifier.predict(myinput)
```

```python
#Visulization
```

```python
#Checking Accuracy of Logistic Regression model
from sklearn.metrics import accuracy_score

#Predict on the test set
y_pred = classifier.predict(x_test)

#Calculate accuracy
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy of the Logistic Regression Model:{accuracy*100:2f}%")
```

```python
#For User Input
age = float(input("Enter Age:"))
salary = float(input("Enter EstimatedSalary:"))

myinput = pd.DataFrame(sc.transform([[age,salary]]),columns=['Age','EstimatedSalary'])
result = classifier.predict(myinput)

if result[0] == 1:
    print("Prediction:User will purchase plan")
else:
    print("Prediction:User will not purchase plan")
```
