```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Importing and reading dataset
ds = pd.read_csv("Position_Salaries.csv")
ds
```

```python
#Splitting dataset into Independent and Dependent variables
x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
#Fitting Random Forest Regression Model
from sklearn.ensemble import RandomForestRegressor#Here 'ensemble' used as cluster using plot_tree method
reg = RandomForestRegressor(n_estimators=10,random_state=1)
reg.fit(x,y)
```

```python
#Prediction
myinput = pd.DataFrame(data =[[4]],columns=["Level"])
reg.predict(myinput)
#reg.predict([4])#2nd method to prediction
```

```python
#Visulization
from sklearn.tree import plot_tree#Here 'tree' used as cluster using plot_tree method
plot_tree(reg.estimators_[0], filled=True)#Take by default estimator is 0 
plt.show()
```

```python
#Predict for user input
level = float(input("Enter the level:"))
myinput = pd.DataFrame(data=[[level]],columns=["Level"])
result = reg.predict(myinput)
print(f"Predicted Salary for level {level}:₹{result[0]}")
```
